Este proceso de ETL se realiza con la finalidad cruzar todos los datos.

# Extract

```python
coursera_courses = pd.read_csv('_src/coursera_courses.csv', sep='¦', engine='python')
coursera_reviews = pd.read_csv('_src/coursera_reviews.csv', sep='¦', engine='python')
edx_courses = pd.read_csv('_src/edx_courses.csv', sep='¦', engine='python')
udemy_courses = pd.read_csv('_src/udemy_courses.csv', sep='¦', engine='python')
```

# Transform

## Concat

Es de utilidad combinar los datasets en uno debido a que las columnas de interés existen en varios y algunos de los que no, se pueden rellenar con información de la página de las distintas intituciones.

```python
_coursera_courses = coursera_courses.copy()
_coursera_reviews = coursera_reviews.copy()
_edx_courses = edx_courses.copy()
_udemy_courses = udemy_courses.copy()
```

```python
_coursera_courses['dataset'] = 'coursera_courses'
_edx_courses['dataset'] = 'edx_courses'
_udemy_courses['dataset'] = 'udemy_courses'
```

```python
_coursera_courses
```

```python
_coursera_courses.rename(columns={'name':'title'}, inplace=True)
```

```python
_coursera_courses['is_paid'] = False
_coursera_courses['price'] = 0
```

```python
_edx_courses
```

```python
_edx_courses.rename(columns={'n_enrolled':'num_subscribers','Level':'level'}, inplace=True)
```

```python
_edx_courses['is_paid'] = False
_edx_courses['price'] = 0
```

```python
_edx_courses['course_effort_hs_per_week'] = _edx_courses['course_effort_hs_per_week'].apply(lambda x: x[0: x.find('–')]).astype(int)
_edx_courses['course_effort_hs_per_week'] = _edx_courses['course_effort_hs_per_week'] * _edx_courses['course_length_weeks'].astype(int)
_edx_courses = _edx_courses.rename(columns={'course_effort_hs_per_week':'content_duration'}).drop(columns='course_length_weeks')
```

```python
_udemy_courses.sort_values('content_duration',ascending=False)['course_url'][0]
```

```python
_udemy_courses.rename(columns={'course_title':'title', 'url':'course_url'}, inplace=True)
_udemy_courses['certificate_price'] = 0
```

```python
mooc_courses = pd.concat([_edx_courses,_udemy_courses], axis=0)
```

```python
mooc_courses = mooc_courses[['course_id', 'title', 'is_paid', 'price', 'certificate_price',
       'num_subscribers', 'num_reviews', 'num_lectures', 'level',
       'content_duration', 'published_timestamp', 'subject',
       'summary', 'course_type', 'institution', 'instructors', 'language',
       'subtitles', 'course_description',
       'course_syllabus', 'course_url', 'dataset']]
mooc_courses
```

```python
mooc_courses = pd.concat([mooc_courses, _coursera_courses], axis=0)
```
